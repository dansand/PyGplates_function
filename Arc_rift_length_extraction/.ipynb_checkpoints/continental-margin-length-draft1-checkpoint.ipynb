```python
import pygplates
import matplotlib.pyplot as plt
import points_in_polygons
import sys
sys.path.append('/Users/Simon/GIT/gplates-web/django/GWS/utils/')
from create_gpml import create_gpml_regular_long_lat_mesh
import numpy as np
from skimage import measure

%matplotlib inline


basedir = '/Users/Simon/Data/AgeGrids/'
#COBterrane_file = '%s/ConjugatePicks_14Dec/MyCOBTerranes/Global_EarthByte_GeeK07_COB_Terranes_2015_v1/Global_EarthByte_GeeK07_COB_Terranes_2015_v1_polygon_merged2.shp' % basedir
#COBterrane_file = '%s/FeatureData/COB_Terranes/Seton_etal_ESR2012_COB_Terranes_2012.gpmlz' % basedir
COBterrane_file = '/Users/Simon/Data/EByteSVN/EByteDeformingModels/Matthews++_410-0Ma/StaticGeometries/AgeGridInput/Global_EarthByte_GeeK07_COB_Terranes_Matthews_etal.gpml'

#cs = pygplates.FeatureCollection('../pygplates-alpha/Data/Global_EarthByte_230-0Ma_GK07_AREPS_Coastlines.gpmlz')

#rotation_file = '%s/Seton_etal_ESR2012_2012.1.rot' % basedir
#rotation_file = '%s/intertec_interp_2016/Global_EB_250-0Ma_GK07_2015_v1.rot' % basedir
rotation_file = '/Applications/GPlates-2.0.0/SampleData/FeatureCollections/Rotations/Matthews_etal_GPC_2016_410-0Ma_GK07.rot'



cobter = pygplates.FeatureCollection(COBterrane_file)
rotation_model = pygplates.RotationModel(rotation_file)

#for polygon in reconstructed_polygons:
#    for geometry in polygon.get_reconstructed_geometry():
#        for segment in geometry.get_points():
#            print segment


multipoints = create_gpml_regular_long_lat_mesh(1)

for multipoint in multipoints:
    for mp in multipoint.get_all_geometries():
        points = mp.to_lat_lon_point_list()

```

```python
polygons = []
for feature in cobter:
    
    for geom in feature.get_geometries():
        polygon = []
        #print feature.get_geometry()
    
        polygon = feature
        polygon.set_geometry(pygplates.PolygonOnSphere(geom))

        #print polygon.get_geometry()

        polygons.append(polygon)

cobter = pygplates.FeatureCollection(polygons)
#fc.write('pltest.gpmlz')

```

```python
def run_grid_pip(time,points,polygons,rotation_model):
    
    reconstructed_polygons = []
    pygplates.reconstruct(polygons,rotation_model,reconstructed_polygons,time)
    
    rpolygons = []
    for polygon in reconstructed_polygons:
        if polygon.get_reconstructed_geometry():
            rpolygons.append(polygon.get_reconstructed_geometry())

    polygons_containing_points = points_in_polygons.find_polygons(points, rpolygons)

    lat = []
    lon = []
    zval = []
    for pcp,point in zip(polygons_containing_points,points):
        lat.append(point.get_latitude())
        lon.append(point.get_longitude())
        if pcp is not None:
            zval.append(1)
        else:
            zval.append(0)
            
    bi = np.array(zval).reshape(181,361)
    
    return bi
```

```python
plt.figure(figsize=(15,30))

for i,time in enumerate(np.arange(0,406,25)):
    
    bi = run_grid_pip(time,points,cobter,rotation_model)
    
    plt.subplot(10,2,i+1)
    plt.imshow(bi,origin='lower',extent=[0,360,0,180])

    # Find contours at a constant value of 0.8
    contours = measure.find_contours(bi, 0.5)
    
    contour_polygons = []
    for n,cp in enumerate(contours):
        cpf = pygplates.PolygonOnSphere(zip(cp[:,0]-90,cp[:,1]-180))
        contour_polygons.append(cpf)

    parea = 0.
    plen = 0.
    for pg in contour_polygons:
        if pg.get_area()>0.1:
            plen += pg.get_arc_length()*pygplates.Earth.mean_radius_in_kms
            parea += pg.get_area()

    #print parea/(4*np.pi), plen

    for n, contour in enumerate(contours):
        plt.plot(contour[:, 1], contour[:, 0], linewidth=2)
    plt.title('Age = %0.2f Ma, fractional area = %0.3f , length = %d km' % (time,parea/(4*np.pi),plen))

```
